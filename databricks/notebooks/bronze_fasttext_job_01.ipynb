```python
from pyspark.sql.functions import col

raw_path = "abfss://nlp@nlplakeadls001.dfs.core.windows.net/raw/fasttext/ingest_date=2025-10-10/"

# Read raw text
df_bronze = spark.read.text(raw_path)  # column: value

# (Optional) basic trim / filter
df_bronze = df_bronze.filter(col("value").isNotNull())

# Write to a Delta table (job version, to keep separate from DLT)
df_bronze.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("nlp_databricks.default.bronze_fasttext_job")
```
