```python
from pyspark.sql.functions import count

df_silver = spark.table("nlp_databricks.default.silver_fasttext_job")

df_gold = (
    df_silver
    .groupBy("label")
    .agg(count("*").alias("review_count"))
)

df_gold.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("nlp_databricks.default.gold_fasttext_label_counts_job")
```
